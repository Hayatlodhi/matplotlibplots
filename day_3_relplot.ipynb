```python
# Import seaborn
import seaborn as sns

```

```python
# Apply the default theme
sns.set_theme()
```

```python
# Load an example dataset
titanic = sns.load_dataset("titanic")
titanic.head(1)
```

```python
# Create a visualization
sns.relplot(data=titanic, x="fare", y="age", col="who", hue="who", style="alone", size="who")
```

```python
# Create a visualization
sns.relplot(data=titanic, x="fare", y="age", col="survived", hue="who", style="who", size="who")
```
