```python
# plot
from matplotlib.pyplot import figure
fig, ax = plt.subplots()
ax.bar(x, y,width=1, edgecolor="white", linewidth=0.7)
ax.set(xlim=(0, 8),xticks=np.arange(1,8), yticks=np.arange(1, 8),
 ylim=(0, 8))
plt.show()
```
